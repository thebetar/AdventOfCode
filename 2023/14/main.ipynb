```python
def read_data(input_data):
    data = []
    for line in input_data.strip().split('\n'):
        data.append(list(line))
    return data
```

```python
def roll_rocks(data, direction):
    vertical_update = 0
    horizontal_update = 0

    data_height = len(data)
    data_width = len(data[0])

    vertical_range = range(data_height)
    horizontal_range = range(data_width)

    if direction == 'N':
        vertical_update = -1
        vertical_range = range(1, data_height)
    elif direction == 'S':
        vertical_update = 1
        vertical_range = range(0, data_height - 1)
    elif direction == 'W':
        horizontal_update = -1
        horizontal_range = range(1, data_width)
    elif direction == 'E':
        horizontal_update = 1
        horizontal_range = range(0, data_width - 1)

    # Roll all rows up till they reach another rock or the edge
    while True:
        changes = 0

        for i in vertical_range:
            for j in horizontal_range:
                if data[i][j] == 'O' and data[i + vertical_update][j + horizontal_update] == '.':
                    data[i + vertical_update][j + horizontal_update] = 'O'
                    data[i][j] = '.'
                    changes += 1

        if changes == 0:
            break


    return data
```

```python
def calculate_score(data):
    score = 0

    height = len(data)

    for i in range(len(data)):
        for j in range(len(data[0])):
            if data[i][j] == 'O':
                score += height - i

    return score
```

```python
with open('input.txt', 'r') as f:
    data = read_data(f.read())
    data = roll_rocks(data, 'N')
    data = calculate_score(data)

    print(data)
```

```python
def fast_roll_direction(data, direction):
    data_height = len(data)
    data_width = len(data[0])
    
    # Set direction parameters
    if direction == 'N':
        for j in range(data_width):
            for i in range(1, data_height):
                if data[i][j] == 'O':
                    # Roll up as far as possible
                    current_i = i

                    while current_i > 0 and data[current_i-1][j] == '.':
                        data[current_i-1][j] = 'O'
                        data[current_i][j] = '.'
                        current_i -= 1
    
    elif direction == 'S':
        for j in range(data_width):
            for i in range(data_height-2, -1, -1):
                if data[i][j] == 'O':
                    # Roll down as far as possible
                    current_i = i

                    while current_i < data_height-1 and data[current_i+1][j] == '.':
                        data[current_i+1][j] = 'O'
                        data[current_i][j] = '.'
                        current_i += 1
    
    elif direction == 'W':
        for i in range(data_height):
            for j in range(1, data_width):
                if data[i][j] == 'O':
                    # Roll left as far as possible
                    current_j = j

                    while current_j > 0 and data[i][current_j-1] == '.':
                        data[i][current_j-1] = 'O'
                        data[i][current_j] = '.'
                        current_j -= 1
    
    elif direction == 'E':
        for i in range(data_height):
            for j in range(data_width-2, -1, -1):
                if data[i][j] == 'O':
                    # Roll right as far as possible
                    current_j = j
                    
                    while current_j < data_width-1 and data[i][current_j+1] == '.':
                        data[i][current_j+1] = 'O'
                        data[i][current_j] = '.'
                        current_j += 1
    
    return data
```

```python
with open('input.txt', 'r') as f:
    data = read_data(f.read())

    history = []
    CYCLE_COUNT = 1000000000

    for i in range(CYCLE_COUNT):
        data = fast_roll_direction(data, 'N')
        data = fast_roll_direction(data, 'W')
        data = fast_roll_direction(data, 'S')
        data = fast_roll_direction(data, 'E')

        data_str = '\n'.join([''.join(row) for row in data])

        if data_str in history:
            break

        history.append(data_str)

    # Pattern starts rotating from
    cycle_start_idx = history.index(data_str)
    cycle_length = i - cycle_start_idx
    
    # Calculate remaining cycles needed
    remaining_cycles = (CYCLE_COUNT - i - 1) % cycle_length
    
    # Get the final state
    final_state_idx = cycle_start_idx + remaining_cycles
    data = history[final_state_idx]

    # Convert into list of lists
    data = [list(row) for row in data.split('\n')]

    # Calculate score
    score = calculate_score(data)
    print(score)
```
