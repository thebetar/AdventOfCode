```python
with open('input.txt', 'r') as f:
    input_data = f.read().strip().splitlines()
```

```python
def simulate(steps):
    start_position = None

    for row_index, row in enumerate(input_data):
        for col_index, character in enumerate(row):
            if character == 'S':
                start_position = (row_index, col_index)
                break
        
        if start_position:
            break

    current_positions = {start_position}
    size = len(input_data)

    for _ in range(steps):
        next_positions = set()

        for row_index, col_index in current_positions:
            directions = [(0, 1), (0, -1), (1, 0), (-1, 0)]
            
            for delta_row, delta_col in directions:
                new_row = row_index + delta_row
                new_col = col_index + delta_col
                
                if input_data[new_row % size][new_col % size] != '#':
                    next_positions.add((new_row, new_col))

        current_positions = next_positions

    return len(current_positions)

def part_1():
    return simulate(64)

def part_2(steps=26501365):
    size = len(input_data)
    remainder = steps % size
    
    y0 = simulate(remainder)
    y1 = simulate(remainder + size)
    y2 = simulate(remainder + 2 * size)
    
    n = steps // size
    
    b0 = y0
    b1 = y1 - y0
    b2 = y2 - y1
    
    return b0 + b1 * n + (n * (n - 1) // 2) * (b2 - b1)
```

```python
print(f"Part 1: {part_1()}")
print(f"Part 2: {part_2()}")
```
